```python
def dis(n):
    """ 1. Write a Python program to check if the given number is a Disarium Number? """
    temp=n
    num= str(n)
    sum=0
    for i in range(len(num),0,-1):
        
        a= temp%10
        sum+=a **i
        temp= temp//10
    if sum==n:
        return f"{n} is a disorium number"
    else:
        return f"{n} is not a disorium number"    
```

```python
number= int(input("Enter the number: "))
dis(number)
```
