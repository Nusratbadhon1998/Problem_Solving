```python
"""Write a Python Program to Add Two Matrices?"""

row= int(input("Enter the row number for first matrix : "))
col= int(input("Enter the col number for first matrix : "))
mat1=[]
mat2=[]

for i in range(row):
    a=[]
    for j in range(col):
        m= int(input("Elements:  "))
        a.append(m)
    mat1.append(a)
    
print("------------")
print("Matrix one : ")
for i in range(row) :
    
    for j in range(col):
        print(mat1[i][j], end=' ')
    print() 
    
row1= int(input("Enter the row number for second matrix : "))
col1= int(input("Enter the col number for second matrix : "))

for i in range(row1):
    b=[]
    for j in range(col1):
        n= int(input("Elements:  "))
        b.append(n)
    mat2.append(b)
    
print("------------") 
print("Matrix two : ")
for i in range(row1) :
    for j in range(col1):
        print( mat2[i][j] , end=' ')
    print()
    
print("------------") 

if row!=row1 or col !=col1:
    print("Can not be added")
else:
    print("Addition of Matrix One and Matrix Two is: ")


    for i in range(row):
        for j in range(col):
            print(mat1[i][j] + mat2[i][j], end= " ")
        print()
```

```python
"""Write a Python Program to multiply Two Matrices?"""

row= int(input("Enter the row number for first matrix : "))
col= int(input("Enter the col number for first matrix : "))
mat1=[]
mat2=[]

for i in range(row):
    a=[]
    for j in range(col):
        m= int(input("Elements:  "))
        a.append(m)
    mat1.append(a)
    
print("------------")
print("Matrix one : ")
for i in range(row) :
    
    for j in range(col):
        print(mat1[i][j], end=' ')
    print() 
    
row1= int(input("Enter the row number for second matrix : "))
col1= int(input("Enter the col number for second matrix : "))

for i in range(row1):
    b=[]
    for j in range(col1):
        n= int(input("Elements:  "))
        b.append(n)
    mat2.append(b)
    
print("------------") 
print("Matrix two : ")
for i in range(row1) :
    for j in range(col1):
        print( mat2[i][j] , end=' ')
    print()
    
print("------------") 

result=[]

for i in range(row):
    for j in range(col1):
        result.append([0])

if col!= row1:
    print("Can not be multiplied")
    
else:
    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            for k in range(len(mat2)):
                result[i][j] += mat1[i][k] * mat2[k][j]
           
for r in result:
    print(r)


```

```python
"""3. Write a Python Program to Transpose a Matrix?"""

row= int(input("Enter the row number for matrix : "))
col= int(input("Enter the col number for matrix : "))

mat=[]
trans_mat=[]


for i in range(row):
    a=[]
    b=[]
    for j in range(col):
        m= int(input("Elements:  "))
        a.append(m)
        b.append(0)

    mat.append(a)
    trans_mat.append(b)
    
print("------------")
print("Matrix : ")

    
for m in mat:
    print(m)
    

print("After transpose: ")

for k in range(len(mat)):
    for m in range(len(mat[0])):
            trans_mat[m][k]= mat[k][m]
        
for t in trans_mat:
    print(t)
```

```python
"""4. Write a Python Program to Sort Words in Alphabetic Order?"""

s= input("Enter a string: ")

s1= s.lower()

l= list(s1.split(' '))


for i in range(len(l)-1):
    for j in range(len(l)-1-i):
        if ord(l[j][0]) > ord(l[j+1][0]):
            temp= l[j]
            l[j]= l[j+1]
            l[j+1]=temp

new= str(l)
print(f"Unsorted String: {s}")
print(f"String after sorting : {new}")
```
