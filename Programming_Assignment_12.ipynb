```python
"""Write a Python program to Extract Unique values dictionary values?"""


def unique(d):
    
    new=[]
    
    for i in d.values():
        for j in i:
            if j not in new:
                new.append(j)
    return f"Unique values: {new}"

dic = {}

for i in range(int(input())):
    
    key = input("Enter key: ")
    
    n= int(input("How many value you want to append: "))
    
    values=[]
    
    for i in range(n):
    
        value = input("Enter value: ")
        values.append(value)

    dic[key] = values

unique(dic)
```

```python
"""3. Write a Python program to Merging two Dictionaries?"""


def merge(d1,d2):
    
    for i in d2.keys():
        
        d1[i]= d2[i]
        
    return d1
    

d1= {"x":1,"y":2}
d2= {"a":3,"b":4}

merge(d1,d2)

```

```python
"""Write a Python program to sort Python Dictionaries by Key or Value?"""

num= int(input("Enter the range: "))

d= dict()

for i in range(num):
    try:
        key= int(input("Enter integer number: "))
        value= input("Enter string: ")
        d[key]=value
    except Exception as e:
        print("Exception occused as"+ str(e))

n=list(d.keys())
value=list(d.values())

    
for i in range(len(n)-1):
    for j in range(len(n)-1-i):
        if n[j]> n[j+1]:
            temp= n[j]
            n[j]=n[j+1]
            n[j+1]=temp
            
        
sort_dict=dict.fromkeys(n)           

p=0

for i in sort_dict.keys():
    sort_dict[i]=value[p]
    p+=1
print()  
print(f"Dictionary after sorted by key: {sort_dict}")
```
